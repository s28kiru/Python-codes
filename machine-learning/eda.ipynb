# Python script for basic data exploration and basic plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
# Load the boston dataset from sklearn
data = datasets.load_breast_cancer()
```

```python
# Convert the dataset to dataframe
data_df = pd.DataFrame(data.data, columns=data.feature_names)
data_df = data_df.assign(target=pd.Series(data.target))
data_df.describe()
```

```python
# Explore the dataset columns, shape
data_df.info()
```

#### Plot bar plot of categorical variables

```python
def bar_plot(column_name):
    grouped = column_name.groupby(column_name).count()
    bar_df = grouped.to_frame()
    print(bar_df)
    plt.bar(bar_df.index, bar_df.iloc[:,0])
```

```python
# Convert numerical to object datatype
data_df.target = data_df.target.apply(str)
```

```python
bar_plot(data_df.target)
```
