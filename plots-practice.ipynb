```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("data/cast.csv")
```

```python
data.head()
```

```python
# barplot
col_name = data['year']
grouped = col_name.groupby([col_name]).count()
grouped_df = grouped.to_frame()
plt.figure(figsize = (10,15))
plt.bar(grouped_df.index, grouped_df.iloc[:,0])
```

```python
year_col = data.year
year_grouped = year_col.groupby((year_col // 10) * 10).count()
year_df = year_grouped.to_frame()
```

```python
plt.figure(figsize = (15,10))
plt.bar(year_df.index, year_df.iloc[:, 0])
plt.xticks(year_df.index)
```
